## import

```
import tensorflow as tf
```

### import the data set form kersa (MNITS)

```
mnits = tf.keras.datasets.mnist # 28 x 28 images from 0 to 9
```

```
(x_train, y_train), (x_test, y_test) = mnits.load_data()

x_train = tf.keras.utils.normalize(x_train, axis = 1) # scaled the dataset between 0 & 1
x_test = tf.keras.utils.normalize(x_test, axis = 1) # scaled the dataset between 0 & 1

model = tf.keras.models.Sequential() #define the model of neural network will be using
model.add(tf.keras.layers.Flatten()) #model.add(Flatten(inputLayers)) input Layers
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu)) #model.add(Flatten(inputLayers)) 1st wights Layers
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu)) #model.add(Flatten(inputLayers)) 2ed wights Layers
model.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax)) #model.add(Flatten(inputLayers)) output Layers (softmax) for probability distribution

```

```
import matplotlib.pyplot as plt
plt.imshow(x_train[0], cmap = plt.cm.binary)
plt.show()
```

```
print(y_train[0])  #It's 5 
```

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(x_train, y_train, epochs=3)
```

```
val_loss, val_acc = model.evaluate(x_test, y_test)
print('loss :',val_loss)
print('accurcy :',val_acc)
```

```
model.save('epoch_num_reader.model')
```
